```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity





def load_data(file_path, sheet_name):
    return pd.read_excel(file_path, engine='openpyxl', sheet_name=sheet_name)

def calculate_equations(A, C):
    equations = []
    for i in range(A.shape[0]):
        equation = f"{A[i, 0]}x1 + {A[i, 1]}x2 + {A[i, 2]}x3 = {C[i]}"
        equations.append(equation)
    return equations

def calculate_rank(A):
    return np.linalg.matrix_rank(A)

def calculate_cost_vector(A, C):
    # Calculate the pseudo-inverse of matrix A
    A_pseudo_inv = np.linalg.pinv(A)
    
    # Calculate the cost vector using the pseudo-inverse: Cost = A_pseudo_inv * C
    return np.dot(A_pseudo_inv, C)

def mark_category(df):
    # Mark customers as RICH or POOR based on payments
    df['Category'] = np.where(df['Payment (Rs)'] > 200, 'RICH', 'POOR')
    return df

def split_data(df, X_columns, y_column, test_size=0.2, random_state=42):
    # Separate features (X) and target variable (y)
    X = df[X_columns].values
    y = df[y_column].values
    
    # Split the data into training and testing sets
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

def create_knn_classifier(k=3):
    # Create KNN classifier
    return KNeighborsClassifier(n_neighbors=k)

def evaluate_model(classifier, X_test, y_test):
    # Make predictions on the test set
    y_pred = classifier.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print("\nModel Evaluation:")
    print(f"Accuracy: {accuracy:.2f}")
    print("\nConfusion Matrix:")
    print(conf_matrix)
    print("\nClassification Report:")
    print(class_report)


file_path = r"Lab_Session_Data.xlsx"
sheet_name_purchase = 'Purchase data'
sheet_name_stock = 'IRCTC Stock Price'
sheet_name_thyroid = 'thyroid0387_UCI'

# A1: Load data and calculate cost vector
X_purchase = load_data(file_path, sheet_name_purchase)
A_columns = ['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']
C_column = 'Payment (Rs)'
A = X_purchase[A_columns].values
C = X_purchase[C_column].values.reshape(-1, 1)
imputer = SimpleImputer(strategy='mean')
A = imputer.fit_transform(A)
equations = calculate_equations(A, C)
rank_A = calculate_rank(A)
cost_vector = calculate_cost_vector(A, C)

print("\nEquations:")
for equation in equations:
    print(equation)
print("\nDimensionality of vector space of the data:", rank_A)
print("No. of vectors in A and C are as follows:", np.shape(A)[0], "and", np.shape(C)[0])
print("The rank of matrix A is:", rank_A)
cost_columns = ['Cost of Each Product']
cost_df = pd.DataFrame(data=cost_vector, columns=cost_columns)
print("\nCost of Each Product:")
print(cost_df)

# A2: Calculate model vector for predicting the cost of products
model_vector = calculate_cost_vector(A, C)
model_columns = ['Model Vector X']
model_df = pd.DataFrame(data=model_vector, columns=model_columns)
print("\nModel Vector X for Predicting the Cost of Each Product:")
print(model_df)

# A3: Classify customers
X_purchase = mark_category(X_purchase)
X_columns = ['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']
y_column = 'Category'
X_train, X_test, y_train, y_test = split_data(X_purchase, X_columns, y_column)
knn_classifier = create_knn_classifier(k=3)
knn_classifier.fit(X_train, y_train)
evaluate_model(knn_classifier, X_test, y_test)

# A4: Stock price analysis
X_stock = load_data(file_path, sheet_name_stock)
mean_price = np.mean(X_stock['Price'])
variance_price = np.var(X_stock['Price'])
print("\nMean of Price data:", mean_price)
print("Variance of Price data:", variance_price)

wednesday_prices = X_stock[X_stock['Day'] == 'Wed']['Price']
mean_wednesday_price = np.mean(wednesday_prices)
print("\nMean of Wednesday Price data:", mean_wednesday_price)
print("\nComparison of wednesday mean with population price mean:", {mean_price == mean_wednesday_price})

april_prices = X_stock[X_stock['Month'] == 'Apr']['Price']
mean_april_price = np.mean(april_prices)
print("\nMean of April Price data:", mean_april_price)
print("\nComparison of april mean with population price mean:", {mean_price == mean_april_price})

loss_probability = np.mean(X_stock['Chg%'] < 0)
print("\nProbability of making a loss over the stock:", loss_probability)
    
wednesday = X_stock[X_stock['Day'] == 'Wed']
wednesday_profit_probability = wednesday[wednesday['Chg%']>0]['Chg%'].to_numpy()
print("\nProbability of making a profit on Wednesday:", len(wednesday_profit_probability)/len(wednesday))

conditional_profit_probability =  len(wednesday_profit_probability) / len(wednesday)
print("\nConditional probability of making profit, given that today is Wednesday:", conditional_profit_probability)
    
sns.scatterplot(x="Day", y="Chg%", data=X_stock, hue="Day", palette="viridis")

plt.xlabel('Day of the Week')
plt.ylabel('Change in the Stock Price')
plt.show()
    

```

```python
from scipy import stats
i=0
print("Range of numeric attributes are as follows")
for i,atr in enumerate(X_thyroid.columns):
    if datatype[i] == "Numeric":
        X_thyroid[atr] = pd.to_numeric(X_thyroid[atr], errors='coerce')
        print(f'{atr} -> {X_thyroid[atr].min()}-{X_thyroid[atr].max()}')
        z_score = np.abs(stats.zscore(X_thyroid[atr].dropna()))
        outlier_indices = np.where(z_score>3)[0]
        outliers[atr]= len(outlier_indices)
```

```python
i=0
for i,atr in enumerate(X_thyroid.columns):
    if datatype[i] == "Numeric":
        X_thyroid[atr] = pd.to_numeric(X_thyroid[atr], errors='coerce')
        mean[atr]= X_thyroid[atr].mean()
        var[atr]= X_thyroid[atr].var()
        
print("Mean and Variance is as follows")
for i in mean:
    print(i + " : Mean: " + str(mean[i]) + " \n\tVariance: " + str(var[i]))
```

```python

impute_values = {}

for atr in X_thyroid.columns:
    if pd.api.types.is_numeric_dtype(X_thyroid[atr]):
        has_outliers = np.any(z_score > 3)
        if has_outliers:
            # Use median for numeric attributes with outliers
            median_val = X_thyroid[atr].median()
            impute_values[atr] = median_val
            X_thyroid[atr].fillna(median_val, inplace=True)
        else:
            # Use mean for numeric attributes without outliers
            mean_val = X_thyroid[atr].mean()
            impute_values[atr] = mean_val
            X_thyroid[atr].fillna(mean_val, inplace=True)
            
    elif pd.api.types.is_object_dtype(X_thyroid[atr]):
        # For categorical attributes, use mode
        mode_val = X_thyroid[atr].mode()[0]
        impute_values[atr] = mode_val
        X_thyroid[atr].fillna(mode_val, inplace=True)
print("Imputed Values:")
for atr, value in impute_values.items():
    print(f'{atr}: {value}')
```

```python

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

normalized_data = {}

print("Starting normalization process...")

for column in X_thyroid.columns:
    if pd.api.types.is_numeric_dtype(X_thyroid[column]):
        X_thyroid[column] = pd.to_numeric(X_thyroid[column], errors='coerce')
        min_val = X_thyroid[column].min()
        max_val = X_thyroid[column].max()

        if max_val != min_val:
            print(f"Normalizing '{column}' using Min-Max Scaling.")
            X_thyroid[column] = min_max_scaler.fit_transform(X_thyroid[[column]])
        else:
            print(f"Normalizing '{column}' using Standardization.")
            X_thyroid[column] = standard_scaler.fit_transform(X_thyroid[[column]])

        normalized_data[column] = X_thyroid[column]

print("\nNormalized Data Preview:")
print(X_thyroid.head())

print("\nNormalization process complete.")
```

```python

import matplotlib.pyplot as plt



X_thyroid = X_thyroid.reset_index(drop=True)


X_thyroid_numeric = X_thyroid.copy()

for col in X_thyroid_numeric.columns:
    if X_thyroid_numeric[col].dtype == 'object':
        X_thyroid_numeric[col] = LabelEncoder().fit_transform(X_thyroid_numeric[col])

num_observations = 20
X_subset = X_thyroid_numeric.head(num_observations)

# Initialize matrices for JC, SMC, and COS
jc_matrix = np.zeros((num_observations, num_observations))
smc_matrix = np.zeros((num_observations, num_observations))
cos_matrix = np.zeros((num_observations, num_observations))

# Calculate JC, SMC, and COS between pairs of vectors
for i in range(num_observations):
    for j in range(num_observations):
        if i != j:
            vector1 = X_subset.iloc[i].values
            vector2 = X_subset.iloc[j].values
            
            binary_vector1 = vector1.astype(int)
            binary_vector2 = vector2.astype(int)
            f11 = np.sum((binary_vector1 == 1) & (binary_vector2 == 1))
            f10 = np.sum((binary_vector1 == 1) & (binary_vector2 == 0))
            f01 = np.sum((binary_vector1 == 0) & (binary_vector2 == 1))
            denominator_jc = f01 + f10 + f11
            jc_matrix[i, j] = f11 / denominator_jc if denominator_jc > 0 else np.nan

            # Simple Matching Coefficient (SMC)
            f00 = np.sum((binary_vector1 == 0) & (binary_vector2 == 0))
            denominator_smc = f00 + f01 + f10 + f11
            smc_matrix[i, j] = (f11 + f00) / denominator_smc if denominator_smc > 0 else np.nan

            # Cosine Similarity (COS)
            vector1 = vector1.reshape(1, -1)
            vector2 = vector2.reshape(1, -1)
            cos_matrix[i, j] = cosine_similarity(vector1, vector2)[0, 0]

# Plot heatmaps
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.heatmap(jc_matrix, annot=True, ax=axes[0], cmap='coolwarm', vmin=0, vmax=1)
axes[0].set_title('Jaccard Coefficient (JC)')
axes[0].set_xlabel('Observation')
axes[0].set_ylabel('Observation')

sns.heatmap(smc_matrix, annot=True, ax=axes[1], cmap='coolwarm', vmin=0, vmax=1)
axes[1].set_title('Simple Matching Coefficient (SMC)')
axes[1].set_xlabel('Observation')
axes[1].set_ylabel('Observation')

sns.heatmap(cos_matrix, annot=True, ax=axes[2], cmap='coolwarm', vmin=0, vmax=1)
axes[2].set_title('Cosine Similarity (COS)')
axes[2].set_xlabel('Observation')
axes[2].set_ylabel('Observation')

plt.tight_layout()
plt.show()
```
